```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
import csv

# plt.style.use() - For different styles
# print(plt.style.available) - Available Plot styles

plt.style.use('ggplot')

languages = []
people = []
items = pd.read_csv('data.csv')
lang_responses = items['LanguagesWorkedWith']
lang_counter = Counter()

for response in lang_responses:
    lang_counter.update(response.split(';'))

for row in lang_counter.most_common(15):
    languages.append(row[0])
    people.append(row[1])

languages.reverse()
people.reverse()
plt.barh(languages, people) # Expects y axis first since its horizontal bar graph

plt.xlabel('Number of people who use')
plt.title('Most Popular Languages')


plt.savefig('plot_2.png')
plt.show()
```
